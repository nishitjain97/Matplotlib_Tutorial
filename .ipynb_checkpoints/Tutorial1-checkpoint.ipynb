```python
# Load package
import matplotlib.pyplot as plt

# So that we don't have to call plt.show() to view every plot
%matplotlib inline
```

```python
# Plotting point at (1, 1)
plt.scatter(1, 1);
```

```python
# Controlling size of plot
plt.figure(figsize=(15, 15))
plt.scatter(1, 1);
```

```python
# Subplot
for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.text(0.5, 0.5, str((2, 2, i)), ha='center', fontsize=10)
    plt.tight_layout()
```

```python
# Styles and properties
x = [i for i in range(10)]
y = [2 * i for i in range(10)]

plt.plot(x, y, color='green', linewidth=2, linestyle='-');
```

```python
# Marker style
plt.scatter(x, y, marker='o', color='red');
```

```python
# Alpha
plt.plot(x, y, alpha=0.5);
```

```python
# Legend
plt.legend();
```

```python
plt.legend(loc='lower right')
```
